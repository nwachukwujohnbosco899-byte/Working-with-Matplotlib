```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
x= np.arange(0,10)
```

```python
y= x*2
```

```python
plt.plot(x,y);
plt.xlabel('X VIEW')
plt.ylabel ('Y VIEW')
plt.title ('MY FIRST AXIS')
plt.xlim(0,6)
plt.ylim(0,6)
plt.show()
plt.savefig('My New Plot')
```

```python
w= np.linspace(0,10,10)
v= x*4
```

```python
fig=plt.figure()
# this creates an empty figure
axes= fig.add_axes([0,0,0.5,0.5])
axes.plot(w,v)
plt.show()
```

```python
a= np.linspace(0,10,11)
b= a**4
```

```python
fig= plt.figure(dpi=500)
axes1= fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.25,0.25])
#plot axes for axes 1
axes1.plot(a,b)
axes1.set_xlabel('X BIG LABEL')
axes1.set_ylabel('X BIG LABEL')
axes1.set_title( 'BIG TABLE')
axes1.set_xlim(0)
axes1.set_ylim(0)

# plot axes for axes2
axes2.plot(a,a)
axes2.plot(a,a)
axes2.set_xlabel('X SMALL LABEL')
axes2.set_ylabel('X SMALL LABEL')
axes2.set_title( 'SMALL TABLE')
axes2.set_xlim(0)
axes2.set_ylim(0)
plt.show()
```
