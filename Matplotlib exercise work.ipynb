### FIRST QUESTION

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
m= np.linspace(0,10,11)
m
```

```python
c= 3*10**8
```

```python
E= m*c**2
# I am plotting e=mc^2
```

```python
fig=plt.figure(figsize=(4,2),dpi=100)
axes=fig.add_axes([0,1,1,1])
axes.plot(m,E,color='r',linewidth=5)
axes.set_title('E=MC^2')
axes.set_xlabel('Mass in Grams')
axes.set_xlim(0,10)
axes.set_ylabel('Energy in Joules')
plt.show()
```

```python
fig=plt.figure(figsize=(4,2),dpi=100)
axes=fig.add_axes([0,1,1,1])
axes.plot(m,E,color='r',linewidth=5)
axes.set_title('E=MC^2')
axes.set_xlabel('Mass in Grams')
axes.set_xlim(0,10)
axes.set_ylabel('Energy in Joules')
axes.set_yscale('log')
axes.grid(visible=True, which='both', axis='y')
plt.show()
# putting it in a logarithm scale
```

### SECOND QUESTION 

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig=plt.figure()
axes=fig.add_axes([0,1,1,1])
axes.plot(labels,july16_2007,label='july16_2007');
axes.plot(labels,july16_2020,label='july16_2020');
axes.legend(loc=5)
plt.show()
```

```python
fig=plt.figure(dpi=100)
axes=fig.add_axes([0,1,1,1])
axes.plot(labels,july16_2007,label='july16_2007');
axes.plot(labels,july16_2020,label='july16_2020');
axes.legend(bbox_to_anchor=(1.12,0.5),loc='lower center')
plt.show()
```

```python
fig,axes= plt.subplots(nrows=2, ncols=1)
axes[0].plot(labels,july16_2007)
axes[0].set_title('july16_2007')
axes[1].plot(labels,july16_2020)
axes[1].set_title('july16_2020')
fig.subplots_adjust(wspace=0,hspace=0.6)
```

```python
fig, ax1 = plt.subplots()

ax1.plot(labels,july16_2007, lw=1, color="blue")
ax1.set_ylabel(r"area $(m^2)$",fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(labels,july16_2020, lw=1, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")


ax1.spines['left'].set_color('blue')
ax1.spines['left'].set_linewidth(3)
ax2.spines['right'].set_color('red')
ax2.spines['right'].set_linewidth(3)
```
